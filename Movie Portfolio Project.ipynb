```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\Dan\Downloads\movies2.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Checking for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Removing columns containing null values

df.dropna(inplace=True)
df
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Null values removed
# Data types for our columns

df.dtypes
```

```python
# Change data type of collumns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create "Correct Year" column

df['yearcorrect'] = df['released'].astype(str).str[:4]

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].sort_values(ascending=False)

df.head()
```

```python
# Scatterplot with budget vs gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
          
plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "green"}, line_kws={"color":"blue"})
```

```python
# Looking at correlation

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between Budget and Gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
          
plt.xlabel('Moie Features')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Convert Company to numeric values

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
# Visualizing final data using Heatmap


correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
          
plt.xlabel('Moie Features')

plt.ylabel('Budget for Film')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs

df.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs

df.head()
```
